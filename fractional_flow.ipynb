```python
#FRACTIONAL FLOW SIMULATION
#A clear depiction of how the fractional flow curve varies as we change the properties.
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sw = np.linspace(0.01,1,num=50)
```

```python

so = 1-sw
     

krw = sw**3
     

kro = (1-sw)**3
```

```python

plt.plot(sw,krw)
plt.plot(sw,kro)
plt.xlabel('Water Saturation, Sw')
plt.ylabel('Relative permeability,Krel')
plt.legend(['Krw','Kro'])
```

```python
#Now fractional flow part for a horizontal flow
for i in range(3):
  uo = input("\nenter oil viscocity!: ")
  uo = int(uo)
  uw = input("\nenter water viscocity!: ")
  uw = int(uw)
  fw = 1/(1+(uw*kro)/(uo*krw))
  plt.plot(sw,fw)
  plt.xlabel('Water saturaction, Sw')
  plt.ylabel('fractional flow of water, fw(bbl/bbl)')
  plt.legend(['1 CP water','100 CP (polymer) water','1000 CP polymer flood'])
```
